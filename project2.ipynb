```python
from wordcloud import WordCloud, STOPWORDS

from collections import Counter # to get the 10 - 20 pairs

def find_ngrams(input_list, n):
  return zip(*(input_list[i:] for i in range(n)))


def get_most_common_ngrams(lists, n, numberOfpairs):
  count_freq = {}

  # fill dictionary with n-grams as the key and the frequency as the value
  for title in lists:
    # get the bigram
    nGram = find_ngrams(title, 2)
    n_grams_list = [item for item in nGram]

    # compute the frequency 
    for item in n_grams_list:
      if item in count_freq:
        count_freq[item] +=1
      else:
        count_freq[item] = 1

  # get the numberOfpairs most common pairs
  return dict(Counter(count_freq).most_common(numberOfpairs))
```

```python
# get the 20 most common pairs
most_common_20pairs = get_most_common_ngrams(true_titles, 2, 20)
pairs_df = pd.DataFrame.from_dict(most_common_20pairs, orient='index', columns=['pairs'])
pairs_df.reset_index().plot(x = 'index', y = 'pairs', kind='bar', title='True titles', figsize=(10, 6))
plt.show()
```

###Preparing Data

```python
# Prepare data for train
from sklearn.metrics import accuracy_score, f1_score
from sklearn import metrics

# train set
train_data = r'/content/train.csv'
df_train = pd.read_csv(train_data) 
df_train['titles_texts'] = df_train['title'] + " " + df_train['text']

# test set
test_data = r'/content/test.csv'
df_test = pd.read_csv(test_data)
df_test['titles_texts'] = df_test['title'] + " " + df_test['text']

df_test.shape, df_train.shape
```

####Target 

<!-- https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html -->


```python
train_label = np.asarray(df_train['label'])
test_label = np.asarray(df_test['label'])

train_label.shape, test_label.shape
```

####TF-IDF - input

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(ngram_range=(1, 2), min_df=2)

tfidf_train = tfidf_vectorizer.fit_transform(df_train['titles_texts'])
tfidf_test = tfidf_vectorizer.transform(df_test['titles_texts'])
```

####BoW - input

```python
from sklearn.feature_extraction.text import CountVectorizer

countVectorizer = CountVectorizer(ngram_range=(1,2), min_df=2, stop_words='english')

bow_train = countVectorizer.fit_transform(df_train['titles_texts'])
bow_test = countVectorizer.transform(df_test['titles_texts'])
```

####Word2Vec (pretrained model)


```python
from gensim.models import Word2Vec
import gensim.downloader as api

# download the pretrained embeddings
glove_vectors = api.load("glove-wiki-gigaword-100")
#cn_vectors = api.load("conceptnet-numberbatch-17-06-300")
#pre_ft_vectors = api.load('fasttext-wiki-news-subwords-300')
#glove_vectors = api.load('glove-twitter-25')
```

```python
model_w2v = glove_vectors

def document_vector(doc):
    # Create document vectors by averaging word vectors. Remove out-of-vocabulary words
    doc = [word for word in doc if word in model_w2v.vocab]
    return np.mean(model_w2v[doc], axis=0)  

df_train['doc_vectors'] = df_train.titles_texts.apply(document_vector)
df_test['doc_vectors'] = df_test.titles_texts.apply(document_vector)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
print("Printing scores for predictions on the test set.\n")

acc_score_test = accuracy_score(test_label, predictions_LR)     
print ("Average accuracy using Logistic Regression: {}%".format(round(acc_score_test*100,2))) 

f1_score_test = f1_score(test_label, predictions_LR, average='weighted')
print ("Average accuracy using Logistic Regression and f1 measure: {}%".format(round(f1_score_test*100,2)))


fpr, tpr, thresholds = metrics.roc_curve(test_label, predictions_LR, pos_label=1)
print("\n", "Predictions for Logistic Regression")
plt.plot(fpr, tpr)
plt.show()
```

##Logistic Regression

Πείραμε το δεύτερο καλύτερο αποτέλεμσα, γιατι το πρώτο που είναι το SVM με αναπαράσταση BoW έκανε πάρα πολυ χρόνο, οπότε πρακτικά η αναπαράσταση που διαλέξαμε είναι καλύτερη

###Logistic Regression with BOW (Best results)

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

# add stop words like (couldn, t, can, s etc)
stop_words1 = ENGLISH_STOP_WORDS


countVectorizer = CountVectorizer(ngram_range=(1,2), min_df=1, stop_words=stop_words1)
df_test['titles_texts_subject'] = df_test['titles_texts'] + " " + df_test['subject']
df_train['titles_texts_subject'] = df_train['titles_texts'] + " " + df_train['subject']


bow_train = countVectorizer.fit_transform(df_train['titles_texts_subject'])
bow_test = countVectorizer.transform(df_test['titles_texts_subject'])
```

```python
from sklearn.linear_model import LogisticRegression

# Prepare the model
model_LR = LogisticRegression(C=100)

# Fitting our model
model_LR.fit(bow_train, train_label)

# Predict the labels on the test set.
predictions_LR = model_LR.predict(bow_test)
```

```python
print("Printing scores for predictions on the test set:\n")

acc_score_test = accuracy_score(test_label, predictions_LR)     
print ("Average accuracy using Logistic Regression: {}%".format(round(acc_score_test*100,2)))

f1_score_test = f1_score(test_label, predictions_LR, average='weighted')
print ("Average accuracy using Logistic Regression and f1 measure: {}%".format(round(f1_score_test*100,2)))


fpr, tpr, thresholds = metrics.roc_curve(test_label, predictions_LR, pos_label=1)
print("\n", "Predictions for Logistic Regression")
plt.plot(fpr, tpr)
plt.show()
```
